```python
# Complete Colab-ready Script:
# Includes: Autoencoder + PCA + IsolationForest + OneClassSVM anomaly detection
# Dataset: Sample MVTec (hazelnut subset provided)
# Covers: training, evaluation (ROC/PR/F1), visualization, model saving, Flask demo
#
# Author: Vinayak Girish Hotanahalli
# Note: Designed to run on Colab; missing packages will be installed automatically.

# 0. Install required packages and import modules

# Uncomment the install commands below when running on Colab for the first time
# !pip install -q scikit-learn joblib flask ngrok matplotlib seaborn

import os, tarfile, shutil, io, base64, time
from pathlib import Path
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.metrics import (roc_auc_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report,
                             roc_curve)
from sklearn.preprocessing import StandardScaler
import joblib
import cv2
from tensorflow.keras import layers, models
import tensorflow as tf

# Fix random seeds for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# 1. Configuration (edit here)
# If dataset is on Google Drive, mount and set DATASET_PATH accordingly.
# Default: download the sample mvtec.tar (as in assignment)
DOWNLOAD_MVTEC_SAMPLE = True
DATA_TAR_URL = "https://assets.supervisely.com/supervisely-supervisely-assets-public/teams_storage/W/6/pt/ANAGVgKaC62tTrDQWK5JhNP2dd8ynqaTKSM1QdVoAasmTdaLvBwCuW7nCrq9o9lLS2padKnV9QogVGFlEPg7vxEBPIfuFC2Yq7ELNW7xn2t1egLrQPoGpNFJobhh.tar"
MVTEC_TAR_LOCAL = "/content/mvtec.tar"
DATASET_ROOT = Path("/content/mvtec")       # where tar will be extracted
REPO_DATA_DIR = Path("/content/ZS-CLIP-AC-naive/data")  # optional: repo data dir
IMG_SIZE = (128, 128)   # resize images to this size for all models
MAX_IMAGES = None        # None => load all; else limit for speed (e.g., 2000)
AUTOENCODER_EPOCHS = 20
AUTOENCODER_BATCH = 32

# 2. Download & extract dataset (if needed)
if DOWNLOAD_MVTEC_SAMPLE:
    if not Path(MVTEC_TAR_LOCAL).exists():
        print("Downloading sample MVTec tarball...")
        !curl -L -o {MVTEC_TAR_LOCAL} {DATA_TAR_URL}
    else:
        print("Tarball already present, skipping download.")
    if not DATASET_ROOT.exists():
        DATASET_ROOT.mkdir(parents=True, exist_ok=True)
        print("Extracting tarball...")
        with tarfile.open(MVTEC_TAR_LOCAL) as tf:
            tf.extractall(path=DATASET_ROOT)
    else:
        print("Extraction folder already exists:", DATASET_ROOT)
else:
    print("Skipped download; ensure DATASET_ROOT contains dataset.")

# 3. Utility functions
def list_image_files(folder: Path, exts=(".png", ".jpg", ".jpeg")):
    files = [p for p in folder.rglob("*") if p.suffix.lower() in exts]
    return sorted(files)

def read_and_preprocess(p: Path, size=IMG_SIZE):
    img = cv2.imread(str(p))
    if img is None:
        raise ValueError(f"Cannot read image: {p}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, size)
    img = img.astype("float32") / 255.0
    return img

def infer_label_from_filename(fname: str) -> str:
    """
    For hazelnut: hazelnut_<label>_<id>.png
    We'll treat 'good' as normal (0) and others as anomaly classes.
    """
    import re
    m = re.match(r"hazelnut_([A-Za-z0-9\-]+)_", fname)
    if m:
        return m.group(1)
    parts = fname.split("_")
    if len(parts) >= 2:
        return parts[1]
    return "unknown"

# 4. Load images (robust to JSON/meta files)

# Search for hazelnut_* anywhere under DATASET_ROOT; otherwise load all images
all_imgs = list_image_files(DATASET_ROOT)
hazelnut_imgs = [p for p in all_imgs if p.name.startswith("hazelnut_")]
if len(hazelnut_imgs) > 0:
    files_to_use = hazelnut_imgs
else:
    files_to_use = all_imgs

if MAX_IMAGES:
    files_to_use = files_to_use[:MAX_IMAGES]

print(f"Found {len(files_to_use)} images to use (sample path: {files_to_use[0] if len(files_to_use)>0 else 'NONE'})")

# Create arrays and labels
images = []
y_label_names = []
for p in files_to_use:
    try:
        images.append(read_and_preprocess(p))
        y_label_names.append(infer_label_from_filename(p.name))
    except Exception as e:
        print("Skipping file (could not read):", p, e)
images = np.array(images)
print("Loaded images array shape:", images.shape)
# Map labels to binary: 0 good, 1 defective
y_true_binary = np.array([0 if n == "good" else 1 for n in y_label_names])

# 5. Train/Validation split (for autoencoder we train on 'good' images only)
from sklearn.model_selection import train_test_split

# For autoencoder anomaly detection:
# Train autoencoder on only 'good' images (unsupervised normal-data training)
good_idx = np.where(np.array(y_label_names) == "good")[0]
if len(good_idx) < 10:
    # Fallback: if not enough 'good' images, train on all images (less ideal)
    print("Warning: few 'good' images found; training autoencoder on all images.")
    X_train, X_val, idx_train, idx_val = train_test_split(images, range(len(images)), test_size=0.2, random_state=SEED)
else:
    X_good = images[good_idx]
    X_train, X_val = train_test_split(X_good, test_size=0.2, random_state=SEED)
    # For evaluation we will use the entire image set (X_eval) with true labels y_true_binary
X_eval = images
y_eval = y_true_binary
print("Autoencoder train on:", X_train.shape, "validation (normal):", X_val.shape, "eval set:", X_eval.shape)

# 6. Build & train CNN Autoencoder

def build_cnn_autoencoder(input_shape):
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(inp)
    x = layers.MaxPooling2D((2,2), padding='same')(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2), padding='same')(x)
    # Bottleneck
    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)
    encoded = layers.MaxPooling2D((2,2), padding='same')(x)
    # Decoder
    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(encoded)
    x = layers.UpSampling2D((2,2))(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2,2))(x)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2,2))(x)
    decoded = layers.Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)
    model = models.Model(inp, decoded)
    model.compile(optimizer='adam', loss='mse')
    return model

input_shape = (IMG_SIZE[1], IMG_SIZE[0], 3)  # (height,width,channels)
autoencoder = build_cnn_autoencoder(input_shape)
autoencoder.summary()

# Train
history = autoencoder.fit(
    X_train, X_train,
    validation_data=(X_val, X_val),
    epochs=AUTOENCODER_EPOCHS,
    batch_size=AUTOENCODER_BATCH,
    shuffle=True
)

# Save autoencoder
autoencoder_save_path = "/content/autoencoder_model.h5"
autoencoder.save(autoencoder_save_path)
print("Saved autoencoder to", autoencoder_save_path)

# 7. Autoencoder evaluation: compute reconstruction error and metrics

# Reconstruct all eval images
recons = autoencoder.predict(X_eval, batch_size=32)
mse_per_img = np.mean(np.square(X_eval - recons), axis=(1,2,3))

# Choose threshold based on validation normal reconstructions (e.g., 95th percentile)
# If we trained on normal images, we can use errors on X_val to set threshold
if 'X_val' in globals() and len(X_val)>0:
    recons_val = autoencoder.predict(X_val)
    mse_val = np.mean(np.square(X_val - recons_val), axis=(1,2,3))
    threshold = np.percentile(mse_val, 95)
else:
    threshold = np.percentile(mse_per_img, 95)

y_pred_auto = (mse_per_img > threshold).astype(int)

# Metrics
roc_auto = roc_auc_score(y_eval, mse_per_img) if len(np.unique(y_eval))>1 else float("nan")
prec_auto = precision_score(y_eval, y_pred_auto, zero_division=0)
rec_auto = recall_score(y_eval, y_pred_auto, zero_division=0)
f1_auto = f1_score(y_eval, y_pred_auto, zero_division=0)
cm_auto = confusion_matrix(y_eval, y_pred_auto)

print("\nAutoencoder Results:")
print(f"Threshold (95th percentile of normal val): {threshold:.6f}")
print(f"ROC AUC: {roc_auto:.4f}")
print(f"Precision: {prec_auto:.4f}, Recall: {rec_auto:.4f}, F1: {f1_auto:.4f}")
print("Confusion Matrix:\n", cm_auto)
print(classification_report(y_eval, y_pred_auto, zero_division=0))

# Plot ROC curve for autoencoder (if possible)
if len(np.unique(y_eval))>1:
    fpr, tpr, _ = roc_curve(y_eval, mse_per_img)
    plt.figure(figsize=(5,4))
    plt.plot(fpr, tpr, label=f"AUC={roc_auto:.3f}")
    plt.plot([0,1],[0,1],"--", color="gray")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("Autoencoder ROC")
    plt.legend()
    plt.show()

# Visualize a few reconstructions and heatmaps
def plot_reconstructions(idx_list):
    n = len(idx_list)
    plt.figure(figsize=(14, 4))
    for i, idx in enumerate(idx_list):
        orig = (X_eval[idx] * 255).astype(np.uint8)
        rec = (recons[idx] * 255).astype(np.uint8)
        # error map
        err = np.mean(np.abs(X_eval[idx] - recons[idx]), axis=-1)
        # original
        ax = plt.subplot(3, n, i+1)
        plt.imshow(orig); plt.title(f"Orig (label={y_eval[idx]})"); plt.axis("off")
        # recon
        ax = plt.subplot(3, n, i+1+n)
        plt.imshow(rec); plt.title(f"Recon MSE={mse_per_img[idx]:.4f}"); plt.axis("off")
        # heatmap
        ax = plt.subplot(3, n, i+1+2*n)
        plt.imshow(err, cmap='jet'); plt.title("Error heatmap"); plt.axis("off")
    plt.tight_layout()
    plt.show()

# Show samples: choose some anomalies and some normal by score
sorted_idx = np.argsort(mse_per_img)
sample_norm = sorted_idx[:3].tolist()
sample_anom = sorted_idx[-3:].tolist()
plot_reconstructions(sample_norm + sample_anom)

# 8. Baseline: PCA reconstruction error (flatten images)

# Flatten images and standardize for PCA
X_flat = X_eval.reshape((X_eval.shape[0], -1))
scaler = StandardScaler()
X_flat_scaled = scaler.fit_transform(X_flat)

pca = PCA(n_components=0.95, svd_solver='full')  # keep 95% variance
X_pca = pca.fit_transform(X_flat_scaled)
X_recon = pca.inverse_transform(X_pca)
X_recon_orig = scaler.inverse_transform(X_recon)
X_recon_imgs = X_recon_orig.reshape(X_eval.shape)

mse_pca = np.mean(np.square(X_eval - X_recon_imgs), axis=(1,2,3))
threshold_pca = np.percentile(mse_pca, 95)
y_pred_pca = (mse_pca > threshold_pca).astype(int)

roc_pca = roc_auc_score(y_eval, mse_pca) if len(np.unique(y_eval))>1 else float("nan")
prec_pca = precision_score(y_eval, y_pred_pca, zero_division=0)
rec_pca = recall_score(y_eval, y_pred_pca, zero_division=0)
f1_pca = f1_score(y_eval, y_pred_pca, zero_division=0)

print("\nPCA Baseline Results:")
print(f"ROC AUC: {roc_pca:.4f}, Precision: {prec_pca:.4f}, Recall: {rec_pca:.4f}, F1: {f1_pca:.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_eval, y_pred_pca))

# Save PCA model & scaler
joblib.dump(pca, "/content/pca_model.joblib")
joblib.dump(scaler, "/content/pca_scaler.joblib")
print("Saved PCA and scaler.")

# 9. Baseline: IsolationForest (on flattened features)

iso = IsolationForest(n_estimators=100, contamination='auto', random_state=SEED)
iso.fit(X_flat_scaled)  # unsupervised; treat low scores as anomaly
scores_iso = -iso.score_samples(X_flat_scaled)  # higher => more anomalous
threshold_iso = np.percentile(scores_iso, 95)
y_pred_iso = (scores_iso > threshold_iso).astype(int)

roc_iso = roc_auc_score(y_eval, scores_iso) if len(np.unique(y_eval))>1 else float("nan")
prec_iso = precision_score(y_eval, y_pred_iso, zero_division=0)
rec_iso = recall_score(y_eval, y_pred_iso, zero_division=0)
f1_iso = f1_score(y_eval, y_pred_iso, zero_division=0)

print("\nIsolationForest Results:")
print(f"ROC AUC: {roc_iso:.4f}, Precision: {prec_iso:.4f}, Recall: {rec_iso:.4f}, F1: {f1_iso:.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_eval, y_pred_iso))

joblib.dump(iso, "/content/iso_model.joblib")
print("Saved IsolationForest model.")

# 10. Baseline: OneClassSVM (on PCA-reduced features for speed)

# Use PCA features from earlier (X_pca) for OneClassSVM
svm = OneClassSVM(kernel='rbf', gamma='auto', nu=0.05)  # nu ~ expected fraction of outliers
svm.fit(X_pca)  # trained unsupervised on all (or you can train only on good)
scores_svm = -svm.score_samples(X_pca)  # higher => more anomalous
threshold_svm = np.percentile(scores_svm, 95)
y_pred_svm = (scores_svm > threshold_svm).astype(int)

roc_svm = roc_auc_score(y_eval, scores_svm) if len(np.unique(y_eval))>1 else float("nan")
prec_svm = precision_score(y_eval, y_pred_svm, zero_division=0)
rec_svm = recall_score(y_eval, y_pred_svm, zero_division=0)
f1_svm = f1_score(y_eval, y_pred_svm, zero_division=0)

print("\nOneClassSVM Results:")
print(f"ROC AUC: {roc_svm:.4f}, Precision: {prec_svm:.4f}, Recall: {rec_svm:.4f}, F1: {f1_svm:.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_eval, y_pred_svm))

joblib.dump(svm, "/content/ocsvm_model.joblib")
print("Saved OneClassSVM model.")

# 11. Summarize all results in a table and plot heatmap

import pandas as pd
summary = pd.DataFrame({
    "method": ["Autoencoder", "PCA", "IsolationForest", "OneClassSVM"],
    "roc_auc": [roc_auto, roc_pca, roc_iso, roc_svm],
    "precision": [prec_auto, prec_pca, prec_iso, prec_svm],
    "recall": [rec_auto, rec_pca, rec_iso, rec_svm],
    "f1": [f1_auto, f1_pca, f1_iso, f1_svm]
})
print("\nOverall summary:")
display(summary)

plt.figure(figsize=(6,4))
sns.heatmap(summary.set_index("method")[["roc_auc","precision","recall","f1"]].astype(float), annot=True, fmt=".3f", cmap="Blues")
plt.title("Model Comparison")
plt.show()

# 12. Save models and helper functions (already saved with joblib/h5). Provide inference helper.

from PIL import Image
def infer_image_autoencoder(img_path, model, threshold):
    img = read_and_preprocess(Path(img_path), size=IMG_SIZE)
    rec = model.predict(np.expand_dims(img,0))[0]
    mse = float(np.mean(np.square(img - rec)))
    is_anom = bool(mse > threshold)
    return {"mse": mse, "is_anomaly": is_anom}

# Save inference helper (just print path)
print("\nAutoencoder inference example (first eval image):")
example_path = files_to_use[0]
print("Example image:", example_path)
print(infer_image_autoencoder(example_path, autoencoder, threshold))

# 13. Optional: Simple Flask app to serve inference (for demo)

# NOTE: Running a Flask server in Colab requires exposing it (e.g., with ngrok). This code is optional.
FLASK_DEMO = False
if FLASK_DEMO:
    try:
        from flask import Flask, request, jsonify
        app = Flask("anomaly_demo")

        @app.route("/predict", methods=["POST"])
        def predict():
            # expects multipart form 'file'
            f = request.files.get("file")
            if f is None:
                return jsonify({"error":"no file uploaded"}), 400
            img_bytes = f.read()
            img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
            img = img.resize(IMG_SIZE)
            arr = np.array(img).astype("float32") / 255.0
            mse = float(np.mean(np.square(arr - autoencoder.predict(arr[np.newaxis])[0])))
            return jsonify({"mse": mse, "threshold": threshold, "is_anomaly": mse>threshold})

        # run app (in Colab you'll need to use ngrok to expose it)
        print("Starting Flask app on port 5000 (use ngrok to expose)...")
        app.run(host="0.0.0.0", port=5000)
    except Exception as e:
        print("Flask run failed:", e)

# 14. Final conclusion (markdown-style print)

print("\n" + "="*60)
print("Final Summary and Conclusions:")
print("- Trained a convolutional autoencoder on normal hazelnut images and used reconstruction error to detect anomalies.")
print("- Implemented three unsupervised baselines (PCA reconstruction, IsolationForest, OneClassSVM) for comparison.")
print("- Evaluated all methods using ROC-AUC, Precision, Recall, F1. See summary table above.")
print("- Visualized reconstructions and per-pixel error heatmaps to help interpret where defects appear.")
print("- Saved models to disk for reproducibility and exported baseline models (joblib).")
print()
print("Suggested next steps if time/budget permits:")
print("  1) Increase input resolution (e.g., 256x256) to capture finer defects.")
print("  2) Try transfer-learning backbones (EfficientNet) for embedding-based anomaly detection.")
print("  3) Add calibration (temperature scaling) for score thresholds and tune nu/contamination hyperparams.")
print("  4) Create a small web demo (Flask + ngrok) for easy reviewer interaction.")
print("="*60)
```
